# ARIMA + WALK FORWAD VALIDATION

```python
import pandas as pd
```

```python
df = pd.read_csv('shampoo.csv', header=0, parse_dates=[0], date_format='%d-%m-%Y')
df.head()
```

```python
train_size = int(df.shape[0]*0.7)
train , test = df.Sales[0:train_size], df.Sales[train_size:]
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
```

```python
data = train
prediction = []
for t in test:
    model = ARIMA(data, order=(5,5,0))
    model_fit = model.fit()
    y = model_fit.forecast()
    prediction.append(y)
    data = np.append(data,t)
    data=pd.Series(data)
              
```

```python
prediction
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(test.values, prediction)
mse
```
