**UPSAMPLIG AND DOWNSAMPLING**

```python
import pandas as pd
```

```python
df = pd.read_csv('us-airlines-monthly-aircraft-miles-flown.csv',parse_dates=[0], header=0)
df.head()
```

*DOWNSAMPLING*

```python
Quaterly_df = df.resample('Q' , on = 'Month').mean()
Quaterly_df.head()
```

```python
Yearly_df = df.resample('A' , on = 'Month').sum()
Yearly_df.head()
```

*UPSAMPLING*

```python
Daily_df = df.resample('D' , on = 'Month').mean()
Daily_df.head(40)
```

### To fill the NaN values

- Draw a straight line joining the points (1963-01-01 and 1963-02-01)
- Connect them

```python
interpolated_df = Daily_df.interpolate(method = 'linear')
interpolated_df.head()
```

```python
interpolated_df.plot()
```

```python
spline_interpolate = Daily_df.interpolate(method = 'spline', order=2)
spline_interpolate.head()
```

```python
spline_interpolate.plot()
```
