```python
df = pd.read_csv('daily-total-female-births-CA.csv',header=0,parse_dates=[0])
df.head()
```

- [x] load data

**TO CREATE COLUMNS OF MONTH, YEAR AND DAY FROM THE DATE COLUMN**

```python
df1 = df.copy()
df1['year']=df1['date'].dt.year
df1['month']=df1['date'].dt.month
df1['day']=df1['date'].dt.day
df1.head()
```

- LAG OPERATOR USING 'shift()'
- WINDOW OPERATOR USING 'rolling(window = ).fn()
- 'expanding()' to compare the values upto that particular value

```python
df1['lag1']=df1['births'].shift(1)
df1['lag365']=df1['births'].shift(365)
df1['Roll_mean']=df1['births'].rolling(window = 2).mean()
df1['Roll_max']=df1['births'].rolling(window = 2).max()
df1['Expanding_max']=df1['births'].expanding().max()
```
